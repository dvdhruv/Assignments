Q1. What is Object-Oriented Programming (OOP).

A1. Object-Oriented Programming (OOP) is a programming paradigm based on the concept of objects, which combine data (attributes) and functions (methods) into a single unit. It helps organize code for reusability, scalability, and clarity.

Q2. What is a class in OOP?

A2. A class in OOP is a blueprint or template for creating objects. It defines the attributes (variables) and methods (functions) that the objects created from it will have.

Q3. What is an object in OOP?

A3. In object-oriented programming (OOP), an object is a fundamental unit that represents a real-world or abstract entity, encapsulating both data and the methods that operate on that data

Q4.  What is the difference between abstraction and encapsulation?

A4. Abstraction means hiding the implementation details and showing only the essential features of an object. It focuses on what an object does rather than how it does it.

Encapsulation means wrapping data (variables) and methods (functions) together into a single unit i.e. a class. It helps in data hiding and protects data from unauthorized access.

Q5. What are dunder methods in Python?

A5. Dunder methods (also called magic methods or special methods) in Python are the methods that start and end with double underscores (__). They are used to define special behavior of objects, like how they are created, represented, or compared.

Q6. Explain the concept of inheritance in OOP?

A6. nheritance in Object-Oriented Programming is a concept where one class (child class) can inherit properties and methods from another class (parent class). It helps in code reusability and allows creating a hierarchy of classes.

Q7.  What is polymorphism in OOP?

A7. Polymorphism in OOP means one name, many forms ie the same function or method can behave differently based on the object or data type using it.

Q8. How is encapsulation achieved in Python?

A8. In Python, encapsulation is achieved by restricting access to variables and methods within a class using access modifiers. It helps protect data from being modified directly from outside the class.

Q9. What is a constructor in Python?

A9. A constructor in Python is a special method used to initialize objects when a class is created. It is defined using the __init__() method and is automatically called when an object is created.

Q10. What are class and static methods in Python?

A10. In Python, class methods and static methods are special types of methods defined inside a class that behave differently from regular instance methods.

Class Method : Defined using the @classmethod decorator. Can access or modify class-level variables, but not instance variables.

Static Method : Defined using the @staticmethod decorator. Used for utility functions related to the class but not dependent on class or instance data.

Q11. What is method overloading in Python?

A11. Method Overloading means having multiple methods with the same name but different parameters in the same class. However, Python does not support true method overloading like Java or C++.

Q12. What is method overriding in OOP?

A12. Method Overriding in OOP occurs when a child class defines a method that already exists in its parent class, but with a different implementation. It allows the child class to modify or extend the behavior of the parent class method.

Q13. What is a property decorator in Python?

A13. A property decorator in Python (@property) is used to convert a class method into an attribute, so it can be accessed without parentheses. It helps in encapsulation — controlling how class attributes are accessed, modified, or deleted while keeping syntax simple.

Q14. Why is polymorphism important in OOP?

A14. Polymorphism is important in OOP because it allows the same function or method name to perform different tasks based on the object calling it. This makes code more flexible, reusable, and easier to maintain.

Q15. What is an abstract class in Python?

A15. An abstract class in Python is a blueprint for other classes that cannot be instantiated directly. It provides a base structure that ensures all subclasses follow the same method pattern, supporting abstraction and consistency in OOP.

A16. Advantages are :

Q17. What is the difference between a class variable and an instance variable?

A17. Class Variable : A class variable is a variable that is declared inside the class but outside any method. It is shared by all the objects of the class. It is used when we want a common value for all objects.

Instance Variable : An instance variable is a variable that is declared inside the constructor (__init__ method) using self. It is unique for each object and stores data specific to that particular instance.

Q18. What is multiple inheritance in Python?

A18. Multiple inheritance in Python is a feature that allows a child class to inherit from more than one parent class. It means the derived (child) class can access properties and methods of all parent classes, promoting code reusability.

Q19. Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python?

__repr__() : The __repr__ method in Python is a special method used to compute the "official" string representation of an object, primarily intended for developers and debugging purposes.

Q20.  What is the significance of the ‘super()’ function in Python?

A20. The super() function in Python is used to call methods of the parent (base) class from within a child (derived) class. It allows the child class to reuse and extend the functionality of the parent class without explicitly naming it, which makes the code more maintainable and flexible.

Q21. What is the significance of the __del__ method in Python?

A21. The __del__() method in Python is a destructor method that is automatically called when an object is about to be destroyed or deleted from memory. It is used to perform cleanup operations, such as closing files, releasing memory, or freeing external resources before the object is removed.

Q22. What is the difference between @staticmethod and @classmethod in Python?

Q24.  What is method chaining in Python OOP?

A24. Method chaining in Python OOP means calling multiple methods on the same object in a single line, one after another. It works by making each method return the object itself (self), allowing the next method to be called immediately on that same object.

Q25. What is the purpose of the __call__ method in Python?

Q2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
from it and implement the area() method in both?

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

c = Circle(3)
r = Rectangle(2, 4)


print("Area of Circle:", c.area())
print("Area of Rectangle:", r.area())
```

Q3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute

```python
class Vehicle:
    def __init__(self, v_type):
        self.v_type = v_type

    def show_type(self):
        print("Vehicle Type:", self.v_type)

class Car(Vehicle):
    def __init__(self, v_type, brand):
        super().__init__(v_type)
        self.brand = brand

    def show_brand(self):
        print("Car Brand:", self.brand)


class ElectricCar(Car):
    def __init__(self, v_type, brand, battery):
        super().__init__(v_type, brand)
        self.battery = battery

    def show_details(self):
        print("Electric Car Details:")
        print("Type:", self.v_type)
        print("Brand:", self.brand)
        print("Battery Capacity:", self.battery)


e_car = ElectricCar("Four Wheeler", "BMW", "100 kWh")

e_car.show_type()
e_car.show_brand()
e_car.show_details()
```

Q5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance

```python
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance

    def deposit(self, amount):
        self.__balance += amount
        print(f"{amount} deposited successfully.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"{amount} withdrawn successfully.")
        else:
            print("Insufficient balance")

    def check_balance(self):
        print(f"Current Balance: {self.__balance}")

account = BankAccount(1000)

account.deposit(500)
account.withdraw(200)
account.check_balance()
```

```python
class MathOperations:
  @classmethod
  def add_numbers(cls, a, b):
    return a + b

  @staticmethod
  def subtract_numbers(a, b):
    return a - b

sum_result = MathOperations.add_numbers(10, 5)
print("Addition Result:", sum_result)

sub_result = MathOperations.subtract_numbers(10, 5)
print("Subtraction Result:", sub_result)
```

Q10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(3, 4)
v2 = Vector(1, 2)

result = v1 + v2

print("Vector 1:", v1)
print("Vector 2:", v2)
print("Resultant Vector:", result)
```

Q14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
and hourly rate. Create a derived class Manager that adds a bonus to the salary.

```python
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        base_salary = super().calculate_salary()
        total_salary = base_salary + self.bonus
        return total_salary

emp = Employee("Dhruv", 40, 100)
mgr = Manager("Raju", 40, 100, 5000)

print(f"{emp.name}'s Salary: {emp.calculate_salary()}")
print(f"{mgr.name}'s Salary (with bonus): {mgr.calculate_salary()}")
```

Q16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
implement the sound() method.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Cow Sound")


class Sheep(Animal):
    def sound(self):
        print("Sheep Sound")

cow = Cow()
sheep = Sheep()

cow.sound()
sheep.sound()
```

Q18. Create a class House with attributes address and price. Create a derived class Mansion that adds an
attribute number_of_rooms.

```python
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    def show_details(self):
        print(f"Address: {self.address}")
        print(f"Price: {self.price}")

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def show_details(self):
        super().show_details()
        print(f"Number of Rooms: {self.number_of_rooms}")

h1 = House("123 Area", 5000000)
m1 = Mansion("456 Area", 25000000, 10)

print("House Details:")
h1.show_details()

print("\nMansion Details:")
m1.show_details()

```
