Clustering

Theoretical Questions

Q1. What is unsupervised learning in the context of machine learning?

A1. Unsupervised learning is a type of machine learning where models are trained on unlabeled data without predefined outputs. The algorithm automatically finds patterns or structures in the data. It is mainly used for tasks like clustering and association. Common examples include K-Means and Hierarchical Clustering, which group similar data points together.

Q2 How does K-Means clustering algorithm work?

A2. K-Means clustering works by dividing data into K clusters based on similarity. First, K initial centroids are chosen. Each data point is assigned to the nearest centroid using a distance measure. Then centroids are recalculated as the mean of cluster points. This process repeats until clusters stabilize and centroids stop changing.

Q3. Explain the concept of a dendrogram in hierarchical clustering?

A3. A dendrogram is a tree-like diagram used in hierarchical clustering to show how data points are grouped into clusters. It represents the merging or splitting of clusters at different distances. By cutting the dendrogram at a specific height, we can determine the number of clusters in the dataset.

Q4. What is the main difference between K-Means and Hierarchical Clustering?

A4. The main difference between K-Means and Hierarchical Clustering is that K-Means requires the number of clusters (K) to be specified in advance, while hierarchical clustering does not. Hierarchical clustering builds a tree-like structure (dendrogram) to show cluster relationships, whereas K-Means directly groups data into clusters based on centroid distance.

Q5. What are the advantages of DBSCAN over K-Means?

A5. DBSCAN has several advantages over K-Means. It does not require specifying the number of clusters beforehand and can identify clusters of arbitrary shapes. DBSCAN is also effective at detecting noise and outliers, while K-Means is sensitive to them. This makes DBSCAN more suitable for datasets with irregular cluster structures.

Q6. When would you use Silhouette Score in clustering?

A6. The Silhouette Score is used in clustering to evaluate how well data points fit within their assigned clusters. It measures the similarity of a point to its own cluster compared to other clusters. A higher silhouette score indicates better-defined and well-separated clusters, helping determine the optimal number of clusters.

Q7. What are the limitations of Hierarchical Clustering?

A7. Hierarchical clustering has several limitations. It is computationally expensive and slow for large datasets because it calculates distances between many data points. Once clusters are merged or split, the process cannot be reversed. It is also sensitive to noise and outliers, which can affect the clustering structure.

Q8. Why is feature scaling important in clustering algorithms like K-Means?

A8. Feature scaling is important in clustering algorithms like K-Means because the algorithm uses distance calculations to form clusters. If features have different scales, variables with larger values dominate the distance measurement. Scaling ensures all features contribute equally, improving the accuracy and fairness of clustering results.

Q9. How does DBSCAN identify noise points?

A9. DBSCAN identifies noise points by checking the density of data points around each point. If a point has fewer neighbors than the minimum required (minPts) within a specified radius (epsilon), it is considered a noise or outlier point. These points do not belong to any cluster.

Q10. Define inertia in the context of K-Means.

A10. Inertia in K-Means is the sum of squared distances between each data point and its nearest cluster centroid. It measures how closely the data points are grouped within clusters. A lower inertia value indicates more compact clusters and better clustering performance.

Q11. What is the elbow method in K-Means clustering?

A11. The Elbow Method is used in K-Means clustering to determine the optimal number of clusters. It involves plotting the inertia (within-cluster sum of squares) against different values of K. The point where the graph forms an “elbow” indicates the best number of clusters, where adding more clusters gives minimal improvement.

Q12.  Describe the concept of "density" in DBSCAN?

A12. In DBSCAN, density refers to the number of data points within a specified radius (epsilon) around a point. A region is considered dense if it contains at least a minimum number of points (minPts). Clusters are formed from these dense regions, while points in low-density areas are treated as noise or outliers.

Q13. Can hierarchical clustering be used on categorical data?

A13. Yes, hierarchical clustering can be used on categorical data if an appropriate distance or similarity measure is applied. Instead of numerical distances like Euclidean distance, measures such as Hamming distance or similarity coefficients can be used to compare categorical features and group similar data points into clusters.

Q14. What does a negative Silhouette Score indicate?

A14. A negative Silhouette Score indicates that a data point is likely assigned to the wrong cluster. It means the point is closer to data points in a different cluster than to those in its own cluster. This suggests poor clustering and overlapping clusters in the dataset.

Q15. Explain the term "linkage criteria" in hierarchical clustering?

A15. Linkage criteria in hierarchical clustering define how the distance between two clusters is calculated when deciding which clusters to merge. Common methods include single linkage (nearest points), complete linkage (farthest points), and average linkage (average distance). These criteria influence how clusters are formed during the clustering process.

Q16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

A16. K-Means clustering may perform poorly on data with varying cluster sizes or densities because it assumes clusters are spherical and similar in size. It assigns points based on distance to the nearest centroid, which can cause larger or denser clusters to dominate. As a result, smaller or irregular clusters may be incorrectly grouped.

Q17. What are the core parameters in DBSCAN, and how do they influence clustering?

A17. The core parameters in DBSCAN are epsilon and minPts.
Epsilon defines the radius within which neighboring points are searched. minPts is the minimum number of points required to form a dense region. If a point has at least minPts within ε, it becomes a core point and forms a cluster. Points not meeting this condition may be labeled as noise or outliers.

Q18. How does K-Means++ improve upon standard K-Means initialization?

A18. K-Means++ improves standard K-Means by selecting initial centroids more carefully instead of choosing them randomly. It spreads the starting centroids far apart based on distance from existing centroids. This reduces poor initialization, improves clustering quality, and often leads to faster convergence and more stable results.

Q19. What is agglomerative clustering?

A19. Agglomerative clustering is a type of hierarchical clustering that follows a bottom-up approach. Initially, each data point is treated as its own cluster. Then the closest clusters are gradually merged step by step based on a distance measure until all points form a single cluster or the desired number of clusters is reached.

Q20.  What makes Silhouette Score a better metric than just inertia for model evaluation?

A20. Silhouette Score is better than inertia because it evaluates both cluster cohesion and separation. While inertia only measures how close data points are within the same cluster, Silhouette Score also considers how far they are from other clusters. This provides a more complete measure of clustering quality.

Q21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=1000, centers=4, n_features=2, random_state=1)

model = KMeans(n_clusters=4, random_state=1)
labels = model.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], marker='x')
plt.show()
```

Q29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

Q31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=1000, centers=3, cluster_std=2.5, random_state=1)

model = KMeans(n_clusters=3, random_state=1)
model.fit(X)

x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z)
plt.scatter(X[:,0], X[:,1], c=model.labels_)
plt.show()
```

Q44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import matplotlib.pyplot as plt
import numpy as np

data = load_iris()
X = data.data
y = data.target

model = KMeans(n_clusters=3, random_state=1)
labels = model.fit_predict(X)

sil_vals = silhouette_samples(X, labels)

y_lower = 0
for i in range(3):
    cluster_vals = sil_vals[labels == i]
    cluster_vals.sort()
    size = cluster_vals.shape[0]
    y_upper = y_lower + size
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_vals)
    y_lower = y_upper + 10

plt.xlabel("Silhouette Coefficient")
plt.ylabel("Samples")
plt.show()
```
