Theory Questions

Q1. What is NumPy, and why is it widely used in Python?

A1. NumPy is a Python library used for numerical and scientific computing.
It provides a special data structure called ndarray (N-dimensional array) that allows fast storage and operations on large numerical datasets.

NumPy is popular because:

1. Very fast : NumPy arrays are implemented in C, so they perform mathematical operations much faster than Python lists.

2. Handles multidimensional data : Support 1D , 2D ,3D+ arrays (tensors for ML & scientific use).

3. Rich mathematical functions : Linear algebra, trigonometry, statistics, random numbers, etc. — all built-in.

4. Broadcasting : We can perform operations on arrays of different shapes without writing loops.

Q2. How does broadcasting work in NumPy?

A2. Broadcasting is a mechanism that allows NumPy to perform element-wise operations on arrays of different shapes by automatically expanding the smaller array without actually copying data.

Q3. What is a Pandas DataFrame?

A3. A Pandas DataFrame is a two-dimensional, tabular data structure in Python that looks like a table with rows and columns.

Q4. Explain the use of the groupby() method in Pandas?

A4. groupby() is a method in Pandas used to split data into groups based on one or more columns, and then apply a function (like sum, mean, count, max, min, etc.) on each group.

Q5.  Why is Seaborn preferred for statistical visualizations?

A5. Seaborn is preferred because it is a high-level data visualization library built on top of Matplotlib and provides beautiful, meaningful, and statistical-style visualizations with very little code.

Q6. What are the differences between NumPy arrays and Python lists?

A6. NumPy arrays are homogeneous, meaning all elements must be of the same data type, whereas Python lists can store mixed data types.NumPy arrays are much faster and memory-efficient because they are implemented in C and optimized for numerical operations, while Python lists are slower since they run in pure Python.Mathematical operations can be performed directly on NumPy arrays (vectorized operations), but lists do not support direct arithmetic and require loops.

Q7. What is a heatmap, and when should it be used?

A7. A heatmap is a graphical visualization that uses colors to represent the magnitude of values in a matrix or table. Higher values are usually shown with darker/brighter colors, and lower values with lighter colors.
It helps in quickly identifying patterns, correlations, and intensity variations in data.

Q8. What does the term “vectorized operation” mean in NumPy?

A8. In NumPy, a vectorized operation means performing arithmetic or mathematical operations on entire arrays at once without using explicit loops.

Q9. How does Matplotlib differ from Plotly?

A9. Matplotlib is mainly used for static, publication-quality graphs, whereas Plotly is focused on interactive and web-based visualizations.Matplotlib creates static charts and provides low-level, detailed control over every element of the plot.Plotly creates interactive charts with features like zooming, hovering, and tooltips, making it suitable for dashboards and web applications.

Q10. What is the significance of hierarchical indexing in Pandas?

A10. Hierarchical indexing (also called MultiIndex) in Pandas allows a DataFrame or Series to have multiple levels of index on rows or columns.
This makes it possible to represent high-dimensional data in a 2-dimensional table and enables more powerful data grouping, reshaping, and slicing operations.

Q11. What is the role of Seaborn’s pairplot() function?

A11. The pairplot() function in Seaborn is used to visualize pairwise relationships between multiple numerical features in a dataset.
It automatically creates a grid of scatterplots for each feature combination, and displays the distribution of each individual variable on the diagonal.

Q12. What is the purpose of the describe() function in Pandas?

A12. The describe() function in Pandas is used to generate summary statistics for numerical (and optionally categorical) columns in a DataFrame.
It provides an overview of key statistical measures, helping to quickly understand the distribution and central tendency of the data.

Q13. Why is handling missing data important in Pandas?

A13. Handling missing data is important because incomplete or null values can negatively affect data analysis and machine learning results.
If missing values are not treated properly, they can lead to incorrect calculations, biased insights, unexpected errors, and unreliable model predictions.

Q14. What are the benefits of using Plotly for data visualization?

A14. Plotly is widely used because it enables the creation of interactive, web-based visualizations with very little code. Unlike traditional static libraries, Plotly provides rich interactivity, making data exploration more intuitive and engaging.

Q15. How does NumPy handle multidimensional arrays?

A15. NumPy handles multidimensional data using a special data structure called the ndarray (N-Dimensional Array).
The ndarray can represent vectors (1D), matrices (2D), and tensors (3D or higher) in a single unified format, allowing fast mathematical operations across all dimensions.

Q16. What is the role of Bokeh in data visualization?

A16. Bokeh is a Python library used for creating interactive, web-based data visualizations.
Its main role is to allow developers to build dynamic, browser-ready charts and dashboards that support real-time updates and user interaction.

Q17. Explain the difference between apply() and map() in Pandas?

A17. map() works only on Series (single column) and performs element-wise transformation. apply() works on both Series and DataFrames and can apply a function row-wise or column-wise, making it more flexible.

Q18. What are some advanced features of NumPy?

A18. Features are:

1. Broadcasting : Allows arithmetic operations between arrays of different shapes without writing loops.

2. Vectorized Operations : Performs mathematical operations on entire arrays at once using optimized C code, giving very fast performance.

3. Multidimensional Arrays : Supports vectors, matrices, and higher-dimensional tensors with efficient memory layout.

4. Universal Functions : Built-in fast mathematical functions (like sin, log, exp, etc.) that work element-wise on arrays.

Q19. How does Pandas simplify time series analysis?

A19. Pandas provides built-in tools specifically designed for handling and analyzing time-based data, which makes time series analysis simple and efficient.
It allows dates and timestamps to be treated as index values, enabling automatic time-aware operations.

Q20. What is the role of a pivot table in Pandas?

A20. A pivot table in Pandas is used to summarize and reorganize data by turning unique values from one column into headings and applying aggregations (like sum, mean, count, etc.) over another column.
It helps in condensing large datasets into meaningful insights by grouping and aggregating data across multiple dimensions.

Q21. Why is NumPy’s array slicing faster than Python’s list slicing?

A21. NumPy slicing is faster because NumPy arrays are stored in contiguous memory and slicing returns a view without copying data.
Python lists store separate objects in scattered memory, so slicing requires copying elements, making it slower and less efficient.

Q22. What are some common use cases for Seaborn?

A22. Seaborn is widely used for statistical data visualization in Python.
It makes it easy to create informative and attractive charts that help understand data patterns, distributions, and relationships.

Q1. How do you create a 2D NumPy array and calculate the sum of each row?

```python
import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])

row_sum = arr.sum(axis=1)
print("Row-wise Sum:")
print(row_sum)
```

Q2. Write a Pandas script to find the mean of a specific column in a DataFrame?

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Dhruv', 'Ajay', 'Raj'],
    'Marks': [85, 92, 78]
})


mean_marks = df['Marks'].mean()

print("Mean of Marks column:", mean_marks)
```

Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Math': [85, 92, 78, 90, 88],
    'Science': [80, 95, 75, 85, 82],
    'English': [78, 88, 82, 91, 84]
})

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

Q6.  Create a DataFrame and add a new column based on an existing column.

Q7. Write a program to perform element-wise multiplication of two NumPy arrays.

Q8. Create a line plot with multiple lines using Matplotlib.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 15, 20, 25, 30]   # First line
y2 = [5, 12, 18, 22, 28]    # Second line
y3 = [2, 8, 14, 20, 26]     # Third line

plt.plot(x, y1, label="A", marker='o')
plt.plot(x, y2, label="B", marker='s')
plt.plot(x, y3, label="C", marker='^')

plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Multiple Lines")
plt.show()
```

Q9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Dhruv', 'Riya', 'Ajay', 'Raj', 'Ram'],
    'Marks': [85, 92, 78, 88, 65]
})

threshold = 80

filtered_df = df[df['Marks'] > threshold]

print("Original DataFrame:\n", df)
print("\nFiltered DataFrame (Marks >", threshold, "):\n", filtered_df)
```

Q11. Perform matrix multiplication using NumPy.

```python
import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

result = A @ B

print("\nMultiplication Result:\n", result)
```

Q13. Create a 3D scatter plot using Plotly.

```python
import plotly.express as px

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 16]
z = [5, 9, 7, 12, 11]

fig = px.scatter_3d(
    x=x,
    y=y,
    z=z,
    title="3D Scatter Plot Example",
    labels={"x": "X Axis", "y": "Y Axis", "z": "Z Axis"},
)

fig.show()
```
