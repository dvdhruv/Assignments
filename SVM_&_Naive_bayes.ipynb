Theoretical

Q1. What is a Support Vector Machine (SVM).

A1. A Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. It works by finding the optimal hyperplane that separates data points of different classes with the maximum margin. The data points closest to the boundary are called support vectors, and they play a key role in defining the decision boundary. SVM is effective in high-dimensional spaces and works well for both linear and non-linear data using kernel functions.

Q2. What is the difference between Hard Margin and Soft Margin SVM.

A2. Hard Margin SVM is used when the data is perfectly linearly separable, meaning no misclassification is allowed. It tries to find a hyperplane that separates the classes with the maximum margin and strictly satisfies all constraints. In contrast, Soft Margin SVM is used when data is not perfectly separable. It allows some misclassification by introducing slack variables and uses a regularization parameter (C) to balance margin maximization and classification error.

Q3.  What is the mathematical intuition behind SVM.

A3. The mathematical intuition behind SVM is to find a hyperplane that maximizes the margin between two classes while minimizing classification error. It does this by solving a convex optimization problem that minimizes (||W||*||W||)/2 , ensuring the decision boundary is as flat as possible. Constraints are added so that data points are correctly classified or penalized using slack variables, making the solution optimal and generalizable.

Q4. What is the role of Lagrange Multipliers in SVM.

A4. Lagrange Multipliers are used in SVM to solve the constrained optimization problem of maximizing the margin while satisfying classification constraints. They help convert the primal optimization problem into its dual form, making it easier to compute.

Q5. What are Support Vectors in SVM.

A5. Support Vectors are the data points that lie closest to the decision boundary (hyperplane) in a Support Vector Machine. These points are critical because they directly determine the position and orientation of the optimal hyperplane.

Q6. What is a Support Vector Classifier (SVC).

A6. A Support Vector Classifier (SVC) is a classification version of Support Vector Machine. It finds the optimal hyperplane that separates different classes with maximum margin. SVC can handle both linear and non-linear data using kernel functions like linear, polynomial, and RBF. It is mainly used for binary classification problems.

Q7. What is a Support Vector Regressor (SVR).

A7. A Support Vector Regressor (SVR) is the regression version of Support Vector Machine. It predicts continuous values by fitting a function within a defined error margin called epsilon (ε). Instead of minimizing total error, SVR tries to keep errors within this margin while maintaining model simplicity.

Q8.  What is the Kernel Trick in SVM.

A8. The Kernel Trick in SVM is a technique used to handle non-linear data. It maps the original input data into a higher-dimensional space where it becomes linearly separable. Instead of computing the transformation explicitly, kernel functions like linear, polynomial, and RBF are used to calculate inner products efficiently, making computation faster and practical.

Q9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.

A9. The Linear Kernel is used when data is linearly separable and works well for high-dimensional datasets. It is simple and computationally fast.

The Polynomial Kernel can model curved decision boundaries by introducing polynomial combinations of features, making it suitable for moderately complex data.

The RBF (Radial Basis Function) Kernel handles highly non-linear data by mapping it into higher dimensions and is widely used due to its flexibility and strong performance.

Q10. What is the effect of the C parameter in SVM.

A10. The C parameter in SVM controls the trade-off between maximizing the margin and minimizing classification error. A large C value tries to classify all training points correctly, resulting in a smaller margin and possible overfitting. A small C allows some misclassification, creating a wider margin and better generalization on unseen data.

Q11. What is the role of the Gamma parameter in RBF Kernel SVM.

A11. The Gamma parameter in an RBF Kernel SVM controls how much influence a single training point has on the decision boundary. A high gamma value makes the model focus closely on individual data points, creating a complex boundary and risk of overfitting. A low gamma creates a smoother, more generalized boundary.

Q12.  What is the Naïve Bayes classifier, and why is it called "Naïve".

A12. Naïve Bayes is a supervised classification algorithm based on Bayes’ Theorem. It predicts the class of a data point by calculating probabilities using prior and likelihood values. It is called “Naïve” because it assumes that all features are independent of each other, which is a strong and often unrealistic assumption, but it works well in practice.

Q13. What is Bayes’ Theorem.

A13. Bayes’ Theorem is a probability formula used to update the probability of an event based on new evidence. Formula :

P(A|B) = (P(B|A)*P(A))/P(B)

Q14.  Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes.

A14. Gaussian Naïve Bayes is used for continuous numerical data and assumes features follow a normal distribution.

Multinomial Naïve Bayes is mainly used for discrete count data, such as word frequencies in text classification.

Bernoulli Naïve Bayes works with binary features (0 or 1), indicating presence or absence of a feature, and is commonly used in text problems with binary word occurrence.

Q15. When should you use Gaussian Naïve Bayes over other variants.

A15. Gaussian Naïve Bayes should be used when the features are continuous numerical values and are approximately normally distributed. It is suitable for datasets like medical measurements, sensor data, or exam scores. Unlike Multinomial or Bernoulli variants, it handles real-valued inputs effectively without requiring data to be converted into counts or binary form.

Q16. What are the key assumptions made by Naïve Bayes.

A16. Naïve Bayes makes two main assumptions. First, it assumes conditional independence, meaning all features are independent of each other given the class label. Second, it assumes that the data follows a specific probability distribution depending on the variant used (Gaussian, Multinomial, or Bernoulli). These simplifying assumptions make the model fast and efficient.

Q17. What are the advantages and disadvantages of Naïve Bayes.

A17. Advantages: Naïve Bayes is simple, fast, and works well with high-dimensional data. It performs especially well in text classification and requires less training data. It is computationally efficient and easy to implement.

Disadvantages: It assumes feature independence, which is often unrealistic. It may perform poorly when features are highly correlated and can struggle with complex relationships in data.

Q18.  Why is Naïve Bayes a good choice for text classification.

A18. Naïve Bayes is a good choice for text classification because it handles high-dimensional data efficiently, such as thousands of words as features. It works well with word frequency or presence data using Multinomial or Bernoulli variants. The independence assumption simplifies computation, making it fast, scalable, and effective for spam detection and sentiment analysis.

Q19. Compare SVM and Naïve Bayes for classification tasks.

A19. SVM is a margin-based classifier that finds an optimal hyperplane to separate classes and works well with high-dimensional and complex data, especially using kernels. It focuses on maximizing margin and can handle non-linear boundaries.

Naïve Bayes is a probabilistic classifier based on Bayes’ Theorem and assumes feature independence. It is faster and simpler but may perform poorly when features are highly correlated.

Q20.  How does Laplace Smoothing help in Naïve Bayes.

A20. Laplace Smoothing is used in Naïve Bayes to handle the problem of zero probability. If a feature value does not appear in the training data for a class, its probability becomes zero, which makes the entire prediction zero. Laplace Smoothing adds a small value (usually 1) to all counts, ensuring no probability becomes zero and improving model robustness.

Q24.  Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision
boundary.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

iris = load_iris()
X = iris.data[:, :2]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)

model = SVC(kernel='poly', degree=3)
model.fit(X_train, y_train)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("SVM with Polynomial Kernel - Decision Boundary")
plt.show()
```

Q27. Write a Python program to train an SVM Classifier with different C values and compare the decision
boundaries visually.

```python
from sklearn.datasets import load_iris
from sklearn.svm import SVC

iris = load_iris()
X = iris.data[:, :2]
y = iris.target

C_values = [0.1, 1, 100]

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

plt.figure(figsize=(15, 4))

for i, C in enumerate(C_values):
    model = SVC(kernel='linear', C=C)
    model.fit(X, y)

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.subplot(1, 3, i+1)
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
    plt.title(f"SVM with C = {C}")

plt.tight_layout()
plt.show()
```

Q31. Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C,
gamma, kernel).

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)

svm = SVC()

param_grid = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 0.01, 0.1, 1],
    'kernel': ['linear', 'rbf']
}

grid_search = GridSearchCV(
    estimator=svm,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy'
)

grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy after Tuning:", accuracy)
```

Q32. Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

#Create Imbalanced Dataset
X, y = make_classification(
    n_samples=1000,
    n_features=10,
    n_classes=2,
    weights=[0.9, 0.1],
    random_state=1
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)

model_normal = SVC(kernel='rbf')
model_normal.fit(X_train, y_train)

y_pred_normal = model_normal.predict(X_test)
accuracy_normal = accuracy_score(y_test, y_pred_normal)

model_weighted = SVC(kernel='rbf', class_weight='balanced')
model_weighted.fit(X_train, y_train)

y_pred_weighted = model_weighted.predict(X_test)
accuracy_weighted = accuracy_score(y_test, y_pred_weighted)

print("Accuracy without Class Weight:", accuracy_normal)
print("Accuracy with Class Weight:", accuracy_weighted)
```

Q37. Write a Python program to train an SVM Classifier using Linear, Polynomial, and RBF kernels on the Breast

Q38. Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the

Q40. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and
compare accuracy.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import RFE
from sklearn.preprocessing import StandardScaler

data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

svm_original = SVC(kernel='linear')
svm_original.fit(X_train, y_train)

y_pred_original = svm_original.predict(X_test)
accuracy_original = accuracy_score(y_test, y_pred_original)

selector = RFE(estimator=SVC(kernel='linear'), n_features_to_select=10)
X_train_rfe = selector.fit_transform(X_train, y_train)
X_test_rfe = selector.transform(X_test)

svm_rfe = SVC(kernel='linear')
svm_rfe.fit(X_train_rfe, y_train)

y_pred_rfe = svm_rfe.predict(X_test_rfe)
accuracy_rfe = accuracy_score(y_test, y_pred_rfe)

print("Accuracy without RFE:", accuracy_original)
print("Accuracy with RFE:", accuracy_rfe)
```

Q45. Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC
score.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)

model = GaussianNB()
model.fit(X_train, y_train)

y_prob = model.predict_proba(X_test)[:, 1]

roc_auc = roc_auc_score(y_test, y_prob)
print("ROC-AUC Score:", roc_auc)

fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Naïve Bayes")
plt.show()
```

Q46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, average_precision_score
from sklearn.preprocessing import StandardScaler

data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = SVC(kernel='rbf', probability=True)
model.fit(X_train, y_train)

y_prob = model.predict_proba(X_test)[:, 1]

precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
avg_precision = average_precision_score(y_test, y_prob)

plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title(f"Precision-Recall Curve (AP = {avg_precision:.2f})")
plt.show()
```
