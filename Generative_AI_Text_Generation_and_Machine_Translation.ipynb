Q2. Explain the role of probabilistic modeling in generative models. How do these models differ from discriminative models?

A2. Probabilistic modeling enables generative models to learn the probability distribution of data and generate new samples that resemble the training data. Generative models estimate the joint probability P(X,Y) or the data distribution P(X). In contrast, discriminative models focus on learning the conditional probability P(YвҲЈX) to classify or predict outcomes. Generative models create new data, while discriminative models distinguish between different classes of data.


Q3. What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?

A3. Autoencoders learn to compress input data into a lower-dimensional representation and then reconstruct the original data. However, they are not designed to generate new meaningful samples.

Variational Autoencoders (VAEs) introduce probabilistic latent variables, allowing the model to learn a continuous latent space. In text generation, VAEs can generate new and diverse text samples by sampling from this latent space, making them more suitable for generative tasks than standard autoencoders.


Q4. Describe the working of attention mechanisms in Neural Machine Translation (NMT). Why are they critical?

A4. Attention mechanisms in Neural Machine Translation (NMT) allow the model to focus on the most relevant words in the source sentence while generating each word of the target sentence. Instead of relying on a fixed-length context vector, attention dynamically assigns weights to input words based on their importance.

This improves translation accuracy, especially for long and complex sentences, making attention a critical component of modern NMT systems and Transformer-based models.

Q6. Use the following small text dataset to train a simple Variational Autoencoder (VAE) for text reconstruction:

["The sky is blue", "The sun is bright", "The grass is green", "The night is dark", "The stars are shining"]
1. Preprocess the data (tokenize and pad the sequences).
2. Build a basic VAE model for text reconstruction.
3. Train the model and show how it reconstructs or generates similar sentences.

Include your code, explanation, and sample outputs.


```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

texts = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

# Padding
max_len = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

print("Vocabulary Size:", len(tokenizer.word_index) + 1)
print("Padded Sequences:\n", padded_sequences)
```

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

input_dim = padded_sequences.shape[1]

# Encoder
input_layer = Input(shape=(input_dim,))
encoded = Dense(8, activation='relu')(input_layer)
latent = Dense(2, activation='relu')(encoded)

# Decoder
decoded = Dense(8, activation='relu')(latent)
output_layer = Dense(input_dim, activation='linear')(decoded)

# Model
vae = Model(input_layer, output_layer)

vae.compile(
    optimizer='adam',
    loss='mse'
)

vae.summary()
```

```python
vae.fit(
    padded_sequences,
    padded_sequences,
    epochs=100,
    batch_size=2,
    verbose=1
)
```

```python
reconstructed = vae.predict(padded_sequences)

reconstructed = np.rint(reconstructed).astype(int)

index_word = {v:k for k,v in tokenizer.word_index.items()}

for original, pred in zip(padded_sequences, reconstructed):

    original_sentence = []
    reconstructed_sentence = []

    for token in original:
        if token in index_word:
            original_sentence.append(index_word[token])

    for token in pred:
        if token in index_word:
            reconstructed_sentence.append(index_word[token])

    print("Original:     ", " ".join(original_sentence))
    print("Reconstructed:", " ".join(reconstructed_sentence))
    print()
```

A Variational Autoencoder (VAE) is a generative model that learns a compressed latent representation of input data and reconstructs it through a decoder. Unlike a standard autoencoder, a VAE learns a probability distribution in the latent space, enabling the generation of new samples. In this experiment, text data is tokenized and padded before being fed into the VAE. After training, the model reconstructs the original sentences and can generate new sentences by sampling from the learned latent space.

Q8. Implement a simple attention-based encoder-decoder model for
English-to-Spanish translation using Tensorflow or PyTorch.

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding, LSTM, Dense, Attention
from tensorflow.keras.models import Model

# Sample English-Spanish dataset
english_sentences = [
    "hello",
    "good morning",
    "how are you"
]

spanish_sentences = [
    "hola",
    "buenos dias",
    "como estas"
]

# Tokenization
eng_tokenizer = tf.keras.preprocessing.text.Tokenizer()
spa_tokenizer = tf.keras.preprocessing.text.Tokenizer()

eng_tokenizer.fit_on_texts(english_sentences)
spa_tokenizer.fit_on_texts(spanish_sentences)

eng_seq = eng_tokenizer.texts_to_sequences(english_sentences)
spa_seq = spa_tokenizer.texts_to_sequences(spanish_sentences)

eng_seq = tf.keras.preprocessing.sequence.pad_sequences(
    eng_seq, padding='post'
)

spa_seq = tf.keras.preprocessing.sequence.pad_sequences(
    spa_seq, padding='post'
)

# Parameters
eng_vocab = len(eng_tokenizer.word_index) + 1
spa_vocab = len(spa_tokenizer.word_index) + 1

embedding_dim = 64
units = 128

# Encoder
encoder_inputs = tf.keras.Input(shape=(None,))
encoder_embed = Embedding(eng_vocab, embedding_dim)(encoder_inputs)

encoder_output, state_h, state_c = LSTM(
    units,
    return_sequences=True,
    return_state=True
)(encoder_embed)

# Decoder
decoder_inputs = tf.keras.Input(shape=(None,))
decoder_embed = Embedding(spa_vocab, embedding_dim)(decoder_inputs)

decoder_output, _, _ = LSTM(
    units,
    return_sequences=True,
    return_state=True
)(
    decoder_embed,
    initial_state=[state_h, state_c]
)

# Attention Layer
attention = Attention()([decoder_output, encoder_output])

# Concatenate
concat = tf.keras.layers.Concatenate()(
    [decoder_output, attention]
)

output = Dense(spa_vocab, activation='softmax')(concat)

model = Model(
    [encoder_inputs, decoder_inputs],
    output
)

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy'
)

model.summary()
```

```python
decoder_target = spa_seq.reshape(
    spa_seq.shape[0],
    spa_seq.shape[1],
    1
)

model.fit(
    [eng_seq, spa_seq],
    decoder_target,
    epochs=50,
    verbose=1
)
```
